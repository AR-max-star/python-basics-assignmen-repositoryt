 Q: Explain the concept of type casting in python with example?


 Ans: Type casting is the process of changing one data type to another in Python.Python provides built-in functions for type casting, such as int(), float(), str(), list(), tuple(), dict(), etc.

  
  1.Implicit Typecasting

  
   In implicit casting, Python automatically converts a smaller data type into a larger data type to avoid data loss. This typically happens during arithmetic operations or assignments.

     Example:
       
      
      x = 5       # Integer
      y = 2.5     # Float
      
      result = x + y  # Python automatically converts x to a float

       print(result)    
       print(type(result))   

       Output: 7.5
       Output: <class 'float'>

  2.Explicit Typecasting

  Explicit type casting is done manually using built-in functions like "int()", "float()", "str()", "list()", etc. This is useful when you need to forcefully change a data type.
    
    Example:
       num = "123"  # String
       num_int = int(num)  # Convert string to integer

       print(num_int)       # Output: 123
       print(type(num_int)) # Output: <class 'int'>

Q: Describe the differnt types of loops in python and thier usecases with example?

Ans: Python supports two primary types of loops "for" and "while". loops allow you to repeatedly execute a block of code while a condition is met or for a defined range of values. Each is use for different tasks.

1. for Loop

The for loop iterates over a sequence (e.g., list, tuple, string, range) or other iterable objects. It is used when you know the number of iterations or when iterating over a collection.


 "for" loop Example:
       
       Iterate over a list using for loop

        fruits = ["apple", "banana", "cherry"]
         for fruit in fruits:
           print(fruit)


     To generate some random numbers

      for i in range(1,10):
         print(i)
     we can also use "break" to stop the code   
      
      for i in range(1,10):
         if i==5:
           break
         print(i)

      continue: Skips the current iteration and moves to the next

         for i in range(1, 6):
            if i == 3:
              continue
            print(i)
   
2. while Loop

The while loop executes a block of code as long as a specified condition is True. It is used when the number of iterations is unknown and depends on a condition.
     
      Example 1: Counting Down


       count = 5
       while count > 0:
         print(f"Countdown: {count}")
         count -= 1

         output:
         Countdown: 5
         Countdown: 4
         Countdown: 3
         Countdown: 2
         Countdown: 1

3. Nested Loops

    You can nest one loop inside another. These are useful when working with multi-dimensional data like matrices or tables.

Example:
       
       for i in range(1, 4):
          for j in range(1, 4):  
             print(f"{i} x {j} = {i * j}")
          print()  # Blank line between tables


      output:

      1 x 1 = 1
      1 x 2 = 2
      1  x 3 = 3

      2 x 1 = 2
      2 x 2 = 4
      2 x 3 = 6

       3 x 1 = 3
       3 x 2 = 6
       3 x 3 = 9
